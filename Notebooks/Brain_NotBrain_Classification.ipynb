```python
import matplotlib.pyplot as plt
from numpy import asarray
from PIL import Image

img_data=[]
label_data=[]
for i in range(len(data)):
    image = Image.open(data[i][0])
    image=image.resize((224,224))
    data1 = asarray(image)
    img_data.append(data1)
    label_data.append(data[i][1])

img_data=np.array(img_data)

print(label_data[0])
```

```python
label_data_num=[]
for i in range(len(label_data)):
    if(label_data[i]=='brain'):
        label_data_num.append(1)
    elif(label_data[i]=='not brain'):
        label_data_num.append(0)
        

label_data_num=np.array(label_data_num)
print(type(label_data_num)   )  
```

```python
X_train=img_data[0:500]
```

```python
X_train = X_train.reshape(-1, 224, 224, 1)
```

```python
X_train.shape
```

```python
y_train=label_data_num[0:500]
```

```python
y_train.shape
```

```python
X_test=img_data[500:717]
```

```python
X_test = X_test.reshape(-1, 224, 224, 1)
```

```python
X_test.shape
```

```python
y_test=label_data_num[500:717]
```

```python
y_test.shape
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224,1)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=10)
```
